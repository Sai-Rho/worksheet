```python
#rows having all null values i,e empty rows

empty_rows = data[data.isna().all(axis=1)].index
print(empty_rows)

#These are the empty rows numbers
```

```python
print('total number of empty rows :',len(empty_rows),'\n')
      
print('percentage of empty rows in the dataset',len(empty_rows)*100 / len(data))

data1=data.drop(empty_rows)
data1
```

```python
# Column name having missing or null values

data1.columns[data1.isnull().any()]
```

```python
missing_percentage = data1.isnull().sum() * 100 / len(data1)
missing_percentage
```

```python
#these are the rows having missing values

missing_values = data1[data1.isnull().any(axis=1)]
missing_values
```

```python
missing_percentage_row = data1.isnull().any(axis=1) * 100 / len(data1)
missing_percentage_row
```

```python
missing_percentage1 = missing_percentage > 30
print(missing_percentage1[missing_percentage1])
data2=data1.drop(['BuildingArea','YearBuilt','CouncilArea'],axis=1)
data2      
```

```python
for i in range(len(data2.index)) : 
    print(" Total NaN values in row", i + 1, ":-", 
          data2.iloc[i].isnull().sum(),"and it's percentage :",data2.iloc[i].isnull().sum()*100 / 18) 
```

```python
data2
```

```python
# percentage of missing values still left column wise

missing_percentage2 = data2.isnull().sum() * 100 / len(data2)
print(missing_percentage2)
```

```python
# Here the target column has missing values and going to treat them with fillna 
# Going to fill the missing values with the mean of the price column

data2['Price'].fillna((data2['Price'].mean()), inplace=True)
data2['Price']
```

```python
missing_percentage3 = data2.isnull().sum() * 100 / len(data2)
print(missing_percentage3)
```

There are many ways to treat missing values

1.Ignore the missing values

2.Drop the missing values

3.Case Deletion

4.Imputation by Mean/Mode/Median

5.Regression Methods

6.Multiple Imputation

Here i'm choosing mean method to fill the missing value

```python
data2['Bedroom2'].fillna((data2['Bedroom2'].mean()), inplace=True)
data2['Bathroom'].fillna((data2['Bathroom'].mean()), inplace=True)
data2['Car'].fillna((data2['Car'].mean()), inplace=True)
data2['Landsize'].fillna((data2['Landsize'].mean()), inplace=True)
data2['Lattitude'].fillna((data2['Lattitude'].mean()), inplace=True)
data2['Longtitude'].fillna((data2['Longtitude'].mean()), inplace=True)
data2
```
